```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
punctuation_table = str.maketrans('', '', string.punctuation)


def preprocess_review(text):
    # Lowercase the text
    text = text.lower()

    # Remove punctuation
    text = text.translate(punctuation_table)

    # Tokenize text
    tokens = sent_tokenize(text)

    # Remove stop words and non-alphabetic tokens
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

    # Lemmatize tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return str(tokens)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/data.csv')

# Convert the 'Item' column to a list
data = df['Item']

print(data)

for i in data:
  i=preprocess_review(i)
data
```

```python
import flair
from flair.data import Sentence
import csv

# Load Flair's sentiment model
sentiment_model = flair.models.TextClassifier.load('en-sentiment')

# Example reviews to analyze
reviews = [
    "The battery life is excellent, but the delivery was slow.",
    "I love the display quality; it’s vibrant and clear.",
    "The price is reasonable, but the durability is questionable."
]

# Aspects you want to analyze
aspects = ['battery life', 'delivery', 'display quality', 'price', 'durability']

# Function to classify sentiments for each aspect in the reviews
def classify_aspect_sentiments(reviews, aspects):
    aspect_sentiments = {aspect: [] for aspect in aspects}

    for review in reviews:
        # Analyze sentiment for the whole review first
        sentence = Sentence(review)
        sentiment_model.predict(sentence)
        sentiment = sentence.labels[0]  # Get the sentiment label

        # Check for each aspect in the review
        for aspect in aspects:
            if aspect in review:
                aspect_sentiments[aspect].append(str(sentiment))  # Store the sentiment for the aspect

    return aspect_sentiments

# Classify sentiments
aspect_sentiments = classify_aspect_sentiments(data, aspects)

# Print results
for aspect, sentiments in aspect_sentiments.items():
    print(f"Sentiment for '{aspect}': {sentiments}")
with open('aspect_sentiments.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Aspect", "Sentiment"])  # Write header

    # Write each aspect's sentiments to the CSV
    for aspect, sentiments in aspect_sentiments.items():
        for sentiment in sentiments:
            writer.writerow([aspect, sentiment])

print("Aspect sentiments saved to 'aspect_sentiments.csv'.")
```
